Data Stemming

```python
# Stemming is the process of producing morphological variants of a root/base word.
# Stemming programs are commonly referred to as stemming algorithms or stemmers. 
# For example, A stemming algorithm reduces the words “chocolates”, “chocolatey”, “choco” to the root word, “chocolate”.
# For example, it reduces the word 'भारतीय' to 'भारत'.
```

```python
# Declaring all the hindi suffixes that are used for stemming.
suffixes = {
    1: ["ो", "े", "ू", "ु", "ी", "ि", "ा"],
    2: ["कर", "ाओ", "िए", "ाई", "ाए", "ने", "नी", "ना", "ते", "ीं", "ती", "ता", "ाँ", "ां", "ों", "ें"],
    3: ["ाकर", "ाइए", "ाईं", "ाया", "ेगी", "ेगा", "ोगी", "ोगे", "ाने", "ाना", "ाते", "ाती", "ाता", "तीं", "ाओं", "ाएं", "ुओं", "ुएं", "ुआं"],
    4: ["ाएगी", "ाएगा", "ाओगी", "ाओगे", "एंगी", "ेंगी", "एंगे", "ेंगे", "ूंगी", "ूंगा", "ातीं", "नाओं", "नाएं", "ताओं", "ताएं", "ियाँ", "ियों", "ियां"],
    5: ["ाएंगी", "ाएंगे", "ाऊंगी", "ाऊंगा", "ाइयाँ", "ाइयों", "ाइयां"],
}

# Defining a fucntion to remove the sufix and conver the word to its root word.
def apply_stem(word):
    for L in 5, 4, 3, 2, 1:
        if len(word) > L + 1:
            for suf in suffixes[L]:
                if word.endswith(suf):
                    return word[:-L]
    return word
```

```python
# Saving the index values in a variable for both fake and valid news.
fake_id=list(fake_news.index)
true_id=list(true_news.index)
```

```python
# Applying stemming on the fake_news dataset.
str_temp=""
count=0
for i in list(fake_news):
  count=0
  for j in list(fake_news[i]):
    for words in j.split():
      str_temp+=apply_stem(words)
      str_temp+=" "
    fake_news.loc[fake_id[count],i]=str_temp
    str_temp=""
    count+=1
```

```python
# Applying stemming on the true_news dataset.
str_temp=""
count=0
for i in list(true_news):
  count=0
  for j in list(true_news[i]):
    for words in j.split():
      str_temp+=apply_stem(words)
      str_temp+=" "
    true_news.loc[true_id[count],i]=str_temp
    str_temp=""
    count+=1
```

Removing the stop words from the sentences.

```python
# Stopwords : The words which are generally filtered out before processing a natural language are called stop words. These are actually the most common words in any language (like articles, prepositions, pronouns, conjunctions, etc) and does not add much information to the text.
# For example: 'मैं', 'हमारा', 'उसका' |
```

```python
# Importing the stopwords
stop=open('Data_Approach1\\Hindi_stopwords.txt', encoding='utf8')
hindi_stopwords=[]
for x in stop:
  x = x.replace('\n', '')
  hindi_stopwords.append(x)
```

```python
# Showing some of the stopwords.
print("Hindi Stopwords: ")
hindi_stopwords
```

```python
# Removing the stopwords from the fake news dataset
str_temp=""
count=0
for i in list(fake_news):
  count=0
  for j in list(fake_news[i]):
    for words in j.split():
      if unicode(words) not in hindi_stopwords:
        str_temp+=words
        str_temp+=" "
    fake_news.loc[fake_id[count],i]=str_temp
    str_temp=""
    count+=1
```

```python
str_temp=""
count=0
for i in list(true_news):
  count=0
  for j in list(true_news[i]):
    for words in j.split():
      if unicode(words) not in hindi_stopwords:
        str_temp+=words
        str_temp+=" "
    true_news.loc[true_id[count],i]=str_temp
    str_temp=""
    count+=1
```
