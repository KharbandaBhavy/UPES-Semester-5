```python
import control as ctrl
import matplotlib.pyplot as plt
import pylab as pl
```

#1

```python
# Transfer Function = The TransferFunction class is used to represent systems in transfer function form.
# Step_response = If the system has multiple inputs and/or multiple outputs, the step response is computed for each input/output pair, with all other inputs set to zero. 
# Used here to plot all theoutputs for consecutive inputs.
```

```python
g = ctrl.TransferFunction(1,[1,1,1])
t,y = ctrl.step_response(g)
print("The g polynomial is:")
print(g) 

ax = plt.subplot(111)
ax.plot(t, y)

plt.title("Time Domain")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.plot(t,y)

# Hide the right and top spines
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

plt.show()
```
